Aqui usamos o rename para trocar os nomes, porém apenas retornamos um resultado de um método. Para realmente alterarmos, precisamos atribuir a DF original.

```python
# Agrupando por continente, números de países por continente
df.groupby("Continente")["Pais"].nunique() # Utilizamos o Unique para que os países não se repitam.
```

```python
df.groupby("Continente")["Pais"].unique() # Aqui ao invés de retornamos 'n' que a quantidade que se repete, retornamos apenas o nome dos países do nosso DF
```
